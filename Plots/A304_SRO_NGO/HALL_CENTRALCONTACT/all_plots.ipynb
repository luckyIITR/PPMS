```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Temps = [ 125, 123, 122, 121, 120, 119, 118, 116,150]
colors = [ 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'g','v']
```

```python
plt.figure(figsize=(8,5))
for i in range(len(Temps)):
    file_name = f"./pickle_data/HALL_CENTRALCONTACT_{Temps[i]}K.pkl"
    df = pd.read_pickle(file_name)
    plt.plot(df['B'], df['R'], label = f"{Temps[i]}K")


plt.xlabel('B')
plt.ylabel('R')
plt.title("R VS B")
plt.legend()
plt.show()
```
