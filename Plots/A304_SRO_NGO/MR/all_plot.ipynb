```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
Temps = [125]
```

```python
for i in range(len(Temps)):
    file_name = f"./pickle_data/MR_{Temps[i]}K.pkl"
    df = pd.read_pickle(file_name)
    plt.plot(df['B'], df['MR'], 'o-', label = f"{Temps[i]}K")

plt.xlabel('H')
plt.ylabel('MR(%)')
plt.title("MR")
plt.legend()
plt.show()
```

```python
# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Function to update the plot
def update_plot():
    ax.clear()  # Clear the previous plot
    df = pd.read_pickle("./pickle_data/MR_125K.pkl")
    # Plot the data
    ax.plot(df['B'], df['MR'], 'b-')  # Replace 'x_column' and 'y_column' with your column names
    plt.draw()  # Redraw the plot

    # Update the plot at a given interval (in milliseconds)
interval = 1000  # Update every second
ani = FuncAnimation(fig, update_plot, interval=interval)
plt.show()
```
