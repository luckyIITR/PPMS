```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
Temps = [10, 100, 115, 119, 125, 150, 200, 300]
```

```python
for i in range(len(Temps)):
    file_name = f"./pickle_data/MR_{Temps[i]}K.pkl"
    df = pd.read_pickle(file_name)
    plt.plot(df['B'], df['MR'], label = f"{Temps[i]}K")

plt.xlabel('H')
plt.ylabel('MR(%)')
plt.title("MR")
plt.legend()
plt.show()
```
