```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Temps = [125, 100, 10, 200]
```

```python
for i in range(len(Temps)):
    file_name = f"./pickle_data/MR_{Temps[i]}K.pkl"
    df = pd.read_pickle(file_name)
    plt.plot(df['B_int'], df['MR'], 'o-', label = f"{Temps[i]}K")

plt.xlabel('H')
plt.ylabel('MR(%)')
plt.title("MR")
plt.legend()
plt.show()
```
