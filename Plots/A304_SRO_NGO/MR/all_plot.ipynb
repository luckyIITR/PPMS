```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
Temps = [10,50, 100, 115, 119, 125, 150, 152, 155, 160, 200, 300]
```

```python
for i in range(len(Temps)):
    file_name = f"./pickle_data/MR_{Temps[i]}K.pkl"
    df = pd.read_pickle(file_name)
    plt.plot(df['B'], df['MR'], label = f"{Temps[i]}K")

plt.xlabel('H')
plt.ylabel('MR(%)')
plt.title("MR")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Example DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [2, 4, 6, 8, 10],
    'C': [3, 6, 9, 12, 15],
    'D': [4, 8, 12, 16, 20]
}
df = pd.DataFrame(data)

# Create checkboxes for each column
checkboxes = [widgets.Checkbox(description=column, value=True) for column in df.columns]

# Update the plot based on selected columns
def update_plot(*args):
    selected_columns = [checkbox.description for checkbox in checkboxes if checkbox.value]
    df[selected_columns].plot()
    plt.show()

# Register the update function to handle checkbox changes
for checkbox in checkboxes:
    checkbox.observe(update_plot, 'value')

# Display the checkboxes
display(*checkboxes)

# Initial plot
update_plot()
```
