```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename='C:\\Users\\Admin\\Desktop\\Data\\4_june_2023\\A304_SRO_NGO_Hall_150K_p1T_n1T.csv'
df = pd.read_csv(filename, names=["I", "B","T", "V1", "V2", "V3", "V4",'Rel_value'])
df['multiplier'] = 10**6
df.head()
# NEED TO ADD MULITPLIER IN THE LABVIEW CODE
```

```python
df['R1'] = df['V1']/(df['multiplier']*df['I'])
df['R2'] = df['V2']/(df['multiplier']*df['I'])
df['R3'] = df['V3']/(df['multiplier']*df['I'])
df['R4'] = df['V4']/(df['multiplier']*df['I'])
```

```python
plt.figure(figsize=(18,18))
plt.subplot(2,2,1)
plt.plot(df['B'],df['R1'])
plt.title('R1 vs B')
plt.ylabel('R1')
plt.xlabel('B')
plt.subplot(2,2,2)
plt.plot(df['B'],df['R2'])
plt.title('R2 vs B')
plt.ylabel('R2')
plt.xlabel('B')
plt.subplot(2,2,3)
plt.plot(df['B'],df['R3'])
plt.title('R3 vs B')
plt.ylabel('R3')
plt.xlabel('B')
plt.subplot(2,2,4)
plt.plot(df['B'],df['R4'])
plt.title('R4 vs B')
plt.ylabel('R4')
plt.xlabel('B')
plt.subplots_adjust(hspace = 0.3, wspace = 0.3)
plt.show()
```

```python
df['R12']=(df['V2']-df['V1'])/(df['multiplier']*df['I']*2)
df['R34']=(df['V4']-df['V3'])/(df['multiplier']*df['I']*2)
```

```python
plt.figure(figsize=(20,9))
plt.subplot(1,2,1)
plt.plot(df['B'],df['R12'],'bo-')
plt.title('R12 vs B')
plt.ylabel('R12')
plt.xlabel('B')
plt.subplot(1,2,2)
plt.plot(df['B'],df['R34'],'bo-')
plt.title('R34 vs B')
plt.ylabel('R34')
plt.xlabel('B')
plt.subplots_adjust(hspace = 0, wspace = 0.3)
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['R'] = (df['R12']+df['R34'])/2
plt.plot(df['B'],df['R'],'go-')
plt.xlabel('B (Oe)',fontsize=15)
plt.ylabel('R_avg (Ohms)',fontsize=15)
plt.title('R_avg vs B (Oe)',fontsize=25)
plt.show()
```
