```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
filename='C:\\Users\\Admin\\Desktop\\Data\\8_march_2023\\A304_SRO_NGO_inside_dynacool_1T_Hall_100K.csv'
df = pd.read_csv(filename, names=["I", "B","T", "V1", "V2", "V3", "V4",'Rel_value'])
df['multiplier'] = 10**6
df
# NEED TO ADD MULITPLIER IN THE LABVIEW CODE
```

### Converting B to integer B

```python
def round_fun(a):
    step = 500
    if a > 0 : q = int((a+1)/step)
    else : q = int((a-1)/step)
    return q*step

df['B_int'] = list(map(round_fun,df['B'].values))
df = df[['I', 'B', 'B_int', 'T', 'V1', 'V2', 'V3', 'V4', 'Rel_value', 'multiplier']].copy()
```

```python
df.head()
```

```python
data_temp = df.copy()
data_temp['R1'] = df['V1']/(df['multiplier']*df['I'])
data_temp['R2'] = df['V2']/(df['multiplier']*df['I'])
data_temp['R3'] = df['V3']/(df['multiplier']*df['I'])
data_temp['R4'] = df['V4']/(df['multiplier']*df['I'])
```

```python
plt.figure(figsize=(18,18))
plt.subplot(2,2,1)
plt.plot(data_temp['B'],data_temp['R1'])
plt.title('R1 vs B')
plt.ylabel('R1')
plt.xlabel('B')
plt.subplot(2,2,2)
plt.plot(data_temp['B'],data_temp['R2'])
plt.title('R2 vs B')
plt.ylabel('R2')
plt.xlabel('B')
plt.subplot(2,2,3)
plt.plot(data_temp['B'],data_temp['R3'])
plt.title('R3 vs B')
plt.ylabel('R3')
plt.xlabel('B')
plt.subplot(2,2,4)
plt.plot(data_temp['B'],data_temp['R4'])
plt.title('R4 vs B')
plt.ylabel('R4')
plt.xlabel('B')
plt.subplots_adjust(hspace = 0.3, wspace = 0.3)
plt.show()
```

## Spliting data into two subset (increasing and decreasing of magnetic field)

B_max to B_min --> Decreasing

B_min to B_max --> increasing


```python
B_max = max(df['B_int'])
B_min = min(df['B_int'])
print(B_max, B_min)
```

```python
# finding index of first appearance of B_min
B_min_first_idx = 0
for i in range(len(df)):
    if df.iloc[i].B_int == B_min:
        B_min_first_idx = i
        break
B_min_first_idx
```

```python
# now split it into two data set
df_dec = df.iloc[0:B_min_first_idx+1].copy()
df_inc = df.iloc[B_min_first_idx+1:].copy()
```

```python
df_dec.head()
```

```python
df_inc.head()
```

### Handling Decreasing data

```python

df_dec.drop_duplicates(subset='B_int', keep='first', inplace=True) # remove duplicates
df_dec = pd.merge(df_dec.assign(key=1), df_dec.assign(key=1), on='key').drop('key', axis=1) # cross product

df_dec = df_dec[df_dec['B_int_x'] == -1*df_dec['B_int_y']] # selecting 
df_dec = df_dec[['I_x','B_int_x','T_x','V1_x','V2_x','V3_x','V4_x','V1_y','V2_y','V3_y','V4_y','multiplier_x']].copy() # req coloumns

df_dec.rename({'I_x' : 'I', 'B_int_x' : 'B', 'T_x' : 'T', 'V1_x' : 'V1', 'V2_x' : 'V2', 'V3_x' : 'V3', 'V4_x' : 'V4', 'V1_y' : 'V5', 'V2_y' : 'V6', 'V3_y' : 'V7', 'V4_y' : 'V8', 'multiplier_x' : 'multiplier'}, axis=1, inplace=True)
df_dec.reset_index(inplace = True, drop = True) # resetting index
.head()
```

```python
df_dec
```

### Handling Increasing data

```python

df_inc.drop_duplicates(subset='B_int', keep='first', inplace=True) # remove duplicates
df_inc = pd.merge(df_inc.assign(key=1), df_inc.assign(key=1), on='key').drop('key', axis=1) # cross product

df_inc = df_inc[df_inc['B_int_x'] == -1*df_inc['B_int_y']] # selecting 
df_inc = df_inc[['I_x','B_int_x','T_x','V1_x','V2_x','V3_x','V4_x','V1_y','V2_y','V3_y','V4_y','multiplier_x']].copy() # req coloumns

df_inc.rename({'I_x' : 'I', 'B_int_x' : 'B', 'T_x' : 'T', 'V1_x' : 'V1', 'V2_x' : 'V2', 'V3_x' : 'V3', 'V4_x' : 'V4', 'V1_y' : 'V5', 'V2_y' : 'V6', 'V3_y' : 'V7', 'V4_y' : 'V8', 'multiplier_x' : 'multiplier'}, axis=1, inplace=True)
df_inc.reset_index(inplace = True, drop = True) # resetting index
df_inc.head()
```

## Concat both data

```python
data = pd.concat([df_dec, df_inc])
data.reset_index(inplace = True, drop = True)
```

```python
data
```

```python
data['RHc'] = (data['V2'] - data['V1'] + data['V5'] - data['V6'])/(df['I']*df['multiplier'])
data['RHd'] = (data['V4'] - data['V3'] + data['V7'] - data['V8'])/(df['I']*df['multiplier'])
```

```python
plt.plot(data['B'], data['RHc'])
```

```python
plt.plot(data['B'], data['RHd'])
```

```python
data['R'] = (data['RHc'] + data['RHd'])/2
```

```python
plt.plot(data['B'], data['R'])
```
