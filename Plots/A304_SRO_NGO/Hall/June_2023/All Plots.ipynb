```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
temps = [125, 123, 122, 121, 120, 119, 118, 116]
colors = ['b','g', 'r', 'c', 'm', 'y', 'k', 'g']
```

```python
plt.figure(figsize=(8,5))

for i in range(len(temps)):
    df_inc =  pd.read_pickle(f'./pickle_data/{temps[i]}K_inc.pkl')
    df_dec = pd.read_pickle(f"./pickle_data/{temps[i]}K_dec.pkl")
    plt.plot(df_inc['B'], df_inc['R'], f'{colors[i]}o-', linewidth=0.8, markersize=3, )
    plt.plot(df_dec['B'],df_dec['R'],f'{colors[i]}o-', linewidth=0.8, markersize=3, label = f'{temps[i]}K')
    
plt.title('R vs B')
plt.ylabel('R')
plt.xlabel('B')
plt.legend()
plt.show()
```
