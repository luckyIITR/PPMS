# van der Pauw resistivity measurement resistivity calculations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename='C:\\Users\\Admin\\Desktop\\Analysis\\Data\\2_june_2023\\NaCoO2_2_RT_380_to_2K.csv'
df = pd.read_csv(filename, names=["T_S", "T","B", "I", "V1", "V2", "V3", "V4", "V5", "V6", "V7", "V8"])
df['multiplier'] = 10**6
df.head()

# NEED TO ADD MULITPLIER IN THE LABVIEW CODE
```

```python
df['R1'] = df['V1']/(df['multiplier']*df['I'])
df['R2'] = df['V2']/(df['multiplier']*df['I'])
df['R3'] = df['V3']/(df['multiplier']*df['I'])
df['R4'] = df['V4']/(df['multiplier']*df['I'])
df['R5'] = df['V5']/(df['multiplier']*df['I'])
df['R6'] = df['V6']/(df['multiplier']*df['I'])
df['R7'] = df['V7']/(df['multiplier']*df['I'])
df['R8'] = df['V8']/(df['multiplier']*df['I'])
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,4,1)
plt.plot(df['T'],df['R1'])
plt.title('R1 vs T')
plt.ylabel('R1')
plt.xlabel('T')
plt.subplot(2,4,2)
plt.plot(df['T'],df['R2'])
plt.title('R2 vs T')
plt.ylabel('R2')
plt.xlabel('T')
plt.subplot(2,4,3)
plt.plot(df['T'],df['R3'])
plt.title('R3 vs T')
plt.ylabel('R3')
plt.xlabel('T')
plt.subplot(2,4,4)
plt.plot(df['T'],df['R4'])
plt.title('R4 vs T')
plt.ylabel('R4')
plt.xlabel('T')
plt.subplot(2,4,5)
plt.plot(df['T'],df['R5'])
plt.title('R5 vs T')
plt.ylabel('R5')
plt.xlabel('T')
plt.subplot(2,4,6)
plt.plot(df['T'],df['R6'])
plt.title('R6 vs T')
plt.ylabel('R6')
plt.xlabel('T')
plt.subplot(2,4,7)
plt.plot(df['T'],df['R7'])
plt.title('R7 vs T')
plt.ylabel('R7')
plt.xlabel('T')
plt.subplot(2,4,8)
plt.plot(df['T'],df['R8'])
plt.title('R8 vs T')
plt.ylabel('R8')
plt.xlabel('T')
plt.subplots_adjust(hspace = 0.4, wspace = 0.6)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(df['T'],1*df['R1'],'go',label='R1')
plt.plot(df['T'],-1*df['R2'],label='R2')
plt.plot(df['T'],1*df['R3'],'bo',label='R3')
plt.plot(df['T'],-1*df['R4'],label='R4')
plt.plot(df['T'],1*df['R5'],'ro',label='R5')
plt.plot(df['T'],-1*df['R6'],label='R6')
plt.plot(df['T'],1*df['R7'],'yo',label='R7')
plt.plot(df['T'],-1*df['R8'],label='R8')
plt.title('R_all vs T',fontsize=30)
plt.ylabel('R_all',fontsize=20)
plt.xlabel('T',fontsize=20)
plt.legend()
plt.show()
```
