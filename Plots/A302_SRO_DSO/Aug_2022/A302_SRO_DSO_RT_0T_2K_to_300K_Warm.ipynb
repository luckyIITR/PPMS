# For VDP with Temperature
Use for VDP with temperature variation.
Output plots : <br>
R1 through R8 vs T together <br>
R_12 vs T ( R_12 uses the formula R_12 = (V2-V1)/I ) <br>
R_34 vs T (R_34 uses the formula R_34 = (V4-V3)/I) <br>
R_56 <br>
R_78 <br>
 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename='C:\\Users\\Admin\\Desktop\\Analysis\\Data\\16_Aug\\A302_SRO_DSO_RT_0T_2K_to_300K_Warm.csv'
df = pd.read_csv(filename, names=["T_S", "T","B", "I", "V1", "V2", "V3", "V4", "V5", "V6", "V7", "V8"])
df['multiplier'] = 1
df
# NEED TO ADD MULITPLIER IN THE LABVIEW CODE
```

```python
df['R1'] = df['V1']/(df['multiplier']*df['I'])
df['R2'] = df['V2']/(df['multiplier']*df['I'])
df['R3'] = df['V3']/(df['multiplier']*df['I'])
df['R4'] = df['V4']/(df['multiplier']*df['I'])
df['R5'] = df['V5']/(df['multiplier']*df['I'])
df['R6'] = df['V6']/(df['multiplier']*df['I'])
df['R7'] = df['V7']/(df['multiplier']*df['I'])
df['R8'] = df['V8']/(df['multiplier']*df['I'])
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,4,1)
plt.plot(df['T'],df['R1'])
plt.title('R1 vs T')
plt.ylabel('R1')
plt.xlabel('T')
plt.subplot(2,4,2)
plt.plot(df['T'],df['R2'])
plt.title('R2 vs T')
plt.ylabel('R2')
plt.xlabel('T')
plt.subplot(2,4,3)
plt.plot(df['T'],df['R3'])
plt.title('R3 vs T')
plt.ylabel('R3')
plt.xlabel('T')
plt.subplot(2,4,4)
plt.plot(df['T'],df['R4'])
plt.title('R4 vs T')
plt.ylabel('R4')
plt.xlabel('T')
plt.subplot(2,4,5)
plt.plot(df['T'],df['R5'])
plt.title('R5 vs T')
plt.ylabel('R5')
plt.xlabel('T')
plt.subplot(2,4,6)
plt.plot(df['T'],df['R6'])
plt.title('R6 vs T')
plt.ylabel('R6')
plt.xlabel('T')
plt.subplot(2,4,7)
plt.plot(df['T'],df['R7'])
plt.title('R7 vs T')
plt.ylabel('R7')
plt.xlabel('T')
plt.subplot(2,4,8)
plt.plot(df['T'],df['R8'])
plt.title('R8 vs T')
plt.ylabel('R8')
plt.xlabel('T')
plt.subplots_adjust(hspace = 0.4, wspace = 0.6)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(df['T'],-1*df['R1'],'go',label='R1')
plt.plot(df['T'],1*df['R2'],label='R2')
plt.plot(df['T'],-1*df['R3'],'bo',label='R3')
plt.plot(df['T'],1*df['R4'],label='R4')
plt.plot(df['T'],-1*df['R5'],'ro',label='R5')
plt.plot(df['T'],1*df['R6'],label='R6')
plt.plot(df['T'],-1*df['R7'],'yo',label='R7')
plt.plot(df['T'],1*df['R8'],label='R8')
plt.title('R_all vs T',fontsize=30)
plt.ylabel('R_all',fontsize=20)
plt.xlabel('T',fontsize=20)
plt.legend()
plt.show()
```

```python
df['R12']=-1*(df['V1']-df['V2'])/(df['multiplier']*df['I']*2)
df['R34']=-1*(df['V3']-df['V4'])/(df['multiplier']*df['I']*2)
df['R56']=-1*(df['V5']-df['V6'])/(df['multiplier']*df['I']*2)
df['R78']=-1*(df['V7']-df['V8'])/(df['multiplier']*df['I']*2)
```

```python
plt.figure(figsize=(17,13))
plt.subplot(2,2,1)
plt.plot(df['T'],df['R12'],'bo-')
plt.title('R12 vs T')
plt.ylabel('R12')
plt.xlabel('T')
plt.subplot(2,2,2)
plt.plot(df['T'],df['R34'],'bo-')
plt.title('R34 vs T')
plt.ylabel('R34')
plt.xlabel('T')
plt.subplot(2,2,3)
plt.plot(df['T'],df['R56'],'bo-')
plt.title('R56 vs T')
plt.ylabel('R56')
plt.xlabel('T')
plt.subplot(2,2,4)
plt.plot(df['T'],df['R78'],'bo-')
plt.title('R78 vs T')
plt.ylabel('R78')
plt.xlabel('T')
plt.subplots_adjust(hspace = 0.3, wspace = 0.3)
plt.show()
```

```python
df['R_a']=-1*(df['V1']-df['V2']+df['V5']-df['V6'])/(4*df['multiplier']*df['I'])
df['R_b']=-1*(df['V3']-df['V4']+df['V7']-df['V8'])/(4*df['multiplier']*df['I'])
```

```python
plt.figure(figsize=(30,10))
plt.subplot(1,2,1)
plt.plot(df['T'],df['R_a'],'bo-')
plt.title('R_a vs T',fontsize=30)
plt.ylabel('R_a',fontsize=22)
plt.xlabel('T',fontsize=22)
plt.subplot(1,2,2)
plt.plot(df['T'],df['R_b'],'bo-')
plt.title('R_b vs T',fontsize=30)
plt.ylabel('R_b',fontsize=22)
plt.xlabel('T',fontsize=22)
plt.subplots_adjust(hspace = 0.3, wspace = 0.3)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df['T'],df['R_a'],'go-')
plt.ylabel('R',fontsize=22)
plt.xlabel('T',fontsize=22)
plt.plot(df['T'],df['R_b'],'bo-')
plt.show()
```

```python
df['R']=(df['R_a']+df['R_b'])/2
df['R']
```

```python
plt.figure(figsize=(15,10))
plt.plot(df['T'],df['R'],'go-')
plt.title('R vs T',fontsize=30)
plt.ylabel('R',fontsize=20)
plt.xlabel('T',fontsize=20)
plt.show()
```
